```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
```

### Read Dataset and exploring

```python
# Read Dataset
path = "LinkedInTechJobsDataset.csv"
df = pd.read_csv(path)
df.head(5)
```

```python
# Exploring dataset
df.info()
```

```python
df.describe()
```

We can see that we have a database of 811 records from LinkedIn Tech Jobs, with a total of 40 variables.
What can we observe regarding the required level of the job postings?"

```python
category_count = df['Involvement'].value_counts()

plt.figure(figsize=(8, 8))   
plt.pie(category_count, labels=category_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Involvement')
plt.axis('equal')  
plt.show()
```

We see that the majority of job postings are for mid-senior and entry-level positions.

```python
# obtaining columns
print(df.columns)
```

```python
skills = df[['PYTHON', 'C++', 'JAVA', 'HADOOP', 'SCALA', 'FLASK',
       'PANDAS', 'SPARK', 'NUMPY', 'PHP', 'SQL', 'MYSQL', 'CSS', 'MONGODB',
       'NLTK', 'TENSORFLOW', 'LINUX', 'RUBY', 'JAVASCRIPT', 'DJANGO', 'REACT',
       'REACTJS', 'AI', 'UI', 'TABLEAU', 'NODEJS', 'EXCEL', 'POWER BI',
       'SELENIUM', 'HTML', 'ML']].sum().sort_values(ascending=False)
```

A question that arises when analyzing this dataset is: What are the most demanded skills in technology job postings? This can be visualized in the following chart. It is worth mentioning that this chart represents the total requirements for a specific skill out of a total of 811 records in the dataset.

```python
plt.figure(figsize=(10, 6))   
skills.plot(kind='bar', color='skyblue')

plt.xlabel('Lenguajes y Tecnolog√≠as')
plt.ylabel('Suma')
plt.title('Suma de Uso de Lenguajes y Tecnolog√≠as')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

We can see that from this sample of job postings, the most demanded skills or tools are AI, UI, Excel, SQL, and Java

### üë• Histogram of total number of applicants per job posting

```python

plt.hist(df[['Total_applicants']], bins=10,)
plt.xlabel('Total_applicants')
plt.ylabel('frequency')
plt.title('Histogram of total number of applicants per job posting')
plt.show()
```

From the histogram, it is evident that the majority of job postings have more than 500 applicants. This underscores the importance of having a professional profile that can stand out among other applicants.
